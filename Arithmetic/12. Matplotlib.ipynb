```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Allos visualisation of the plots inside Jupyter Notebook

%matplotlib inline 
```

```python
x = np.linspace(0,5, 11)

y = x ** 2
```

```python
# FUCTIONAL METHOD FOR SHOW PLOTS

plt.plot(x,y)
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Title')
```

```python
# OOP METHOD FOR SHOW PLOTS

fig = plt.figure()

axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y)

axes.set_xlabel('X Lable')
axes.set_ylabel('Y Lable')
axes.set_title('Title')
```
